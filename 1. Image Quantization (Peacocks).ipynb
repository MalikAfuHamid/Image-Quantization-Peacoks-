# 1. Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# 2. Import Picture

```python
img = plt.imread("data/Burung_merak.jpg")
img.shape
```

```python
plt.imshow(img)
```

# 3. Converting Picture to Data 

```python
img = img.reshape(-1, 3)
img
```

```python
len(img)
```

```python
len(np.unique(img, axis=0))
```

# 4. Cluster 503705 colours into 30 colour palette

```python
kmeans = KMeans(n_clusters=30, random_state=42, n_jobs=-1)
labels = kmeans.fit_predict(img)
```

```python
labels
```

```python
palette = kmeans.cluster_centers_.astype(int)
palette
```

```python
quantized_img = palette[labels].reshape(1067, 1600, 3)
quantized_img
```

```python
plt.imshow(quantized_img)
```

```python
len(np.unique(quantized_img, axis=0))
```

`Now, the picture displayed with fewer pixels is not much different from the original picture. And the new picture has lower storage.`
